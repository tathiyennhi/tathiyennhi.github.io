Cu 1:

Hi quy mu PRICEi = 0 + 1SQFTi + 2BEDRMSi + 3BATHSi + i v gii thch  ngha ca cc h s hi quy?

Hi quy mu  cho l:

**PRICEi = 0 + 1SQFTi + 2BEDRMSi + 3BATHSi + i**

Trong :

PRICEi l gi nh (tnh bng nghn ) ca cn nh th i.
SQFTi l din tch nh ca cn nh th i.
BEDRMSi l s phng ng ca cn nh th i.
BATHSi l s phng tm ca cn nh th i.
0, 1, 2 v 3 l cc h s hi quy.
i l sai s ngu nhin.
 ngha ca cc h s hi quy:

0 l h s chn (intercept), i din cho gi nh trung bnh khi tt c cc bin u vo (SQFT, BEDRMS v BATHS) u bng 0. N cho bit gi nh c bn khi khng c din tch, s phng ng v s phng tm.

1 l h s hi quy ca bin SQFT, th hin mi quan h gia din tch nh v gi nh. Mt gi tr dng ca 1 cho bit rng gi nh tng khi din tch nh tng v ngc li.

2 l h s hi quy ca bin BEDRMS, th hin mi quan h gia s phng ng v gi nh. Gi tr dng ca 2 cho bit rng gi nh c xu hng tng khi s phng ng tng.

3 l h s hi quy ca bin BATHS, th hin mi quan h gia s phng tm v gi nh. Gi tr dng ca 3 cho bit rng gi nh c xu hng tng khi s phng tm tng.

Cc h s hi quy cho php c lng mc  nh hng ca tng bin u vo n gi nh. Da trn cc gi tr ca cc h s ny, ta c th a ra d on v gi nh da trn din tch, s phng ng v s phng tm ca cn nh.

Cu 2:
Tnh c lng im gi PRICE trung bnh khi SQFT tng thm 300 v BEDRMS tng thm 1


 tnh c lng im gi PRICE trung bnh khi SQFT tng thm 300 v BEDRMS tng thm 1, ta s dng hi quy mu  cho v thay th gi tr tng ng vo cng thc:

**PRICEi = 0 + 1SQFTi + 2BEDRMSi + 3BATHSi**

Gi s rng bn  tnh c cc h s hi quy 0, 1, 2, 3 t qu trnh c lng m hnh hi quy. Thay SQFTi = SQFT + 300 v BEDRMSi = BEDRMS + 1 vo cng thc, ta c:

PRICE_estimated = 0 + 1(SQFT + 300) + 2(BEDRMS + 1) + 3BATHS
**bold text**
Vi gi tr  bit ca SQFT v BEDRMS, ta c th tnh c lng gi PRICE trung bnh khi SQFT tng thm 300 v BEDRMS tng thm 1.






Cu 3:
Tnh h s xc nh v h s xc nh iu chnh, gii thch  ngha?

H s xc nh (coefficient of determination), thng c k hiu l R-squared, l mt i lng trong hi quy thng k m t mc  gii thch ca m hnh hi quy i vi bin ph thuc. R-squared thng c tnh bng cch so snh s bin thin ca gi tr d on ca m hnh vi s bin thin ca gi tr thc t.

R-squared c gi tr t 0 n 1. Gi tr cng gn 1, m hnh cng tt trong vic gii thch s bin ng ca bin ph thuc. C th:

R-squared = 0: M hnh khng gii thch c bin ng no ca bin ph thuc.
R-squared = 1: M hnh gii thch c ton b bin ng ca bin ph thuc.
H s xc nh iu chnh (adjusted coefficient of determination) c s dng  iu chnh R-squared khi s lng bin c lp trong m hnh thay i. N b sung yu t ca  phc tp ca m hnh vo phn gim do sai s ngu nhin trong m hnh.

 ngha ca R-squared v h s xc nh iu chnh:

R-squared o lng mc  m cc bin u vo trong m hnh (SQFT, BEDRMS, BATHS) gii thch c s bin ng ca gi nh (PRICE). Gi tr R-squared cao (gn 1) cho thy m hnh gii thch tt v c kh nng d on chnh xc gi nh da trn cc bin u vo.
H s xc nh iu chnh iu chnh R-squared  tnh ton mc  gii thch ca m hnh da trn s lng bin c lp v  phc tp ca m hnh. N nh gi kh nng tng qut ha ca m hnh. Mt gi tr cao ca h s xc nh iu chnh cho thy m hnh c kh nng tng qut ha tt hn.
Khi nh gi mt m hnh hi quy, cn xem xt c R-squared v h s xc nh iu chnh  c ci nhn ton din v mc  gii thch v kh nng tng qut ha ca m hnh.








Cu 4:
Tm khong tin cy i vi 1,2 v 3 vi mc  ngha 5%?


 tm khong tin cy cho cc h s 1, 2 v 3 vi mc  ngha 5%, chng ta cn s dng phn phi Student's t v cc gi tr t m hnh hi quy.

Trong m hnh hi quy, cc h s 1, 2 v 3 c c lng t d liu. Vo mt mc  ngha c th (trong trng hp ny, 5%), chng ta xc nh khong tin cy trong  c kh nng cha gi tr thc s ca h s vi xc sut nht nh.

 tnh khong tin cy, chng ta s dng gi tr tng ng vi mc  ngha 5% v bc t do (degrees of freedom) ca m hnh hi quy. Trong trng hp ny, bc t do l (n - k - 1), trong  n l s lng quan st v k l s lng bin u vo (trong trng hp ny, k = 3).

Vi mi h s i, ta tnh khong tin cy theo cng thc:

Khong tin cy = [i - t_critical * SE(i), i + t_critical * SE(i)]

Trong :

i l c lng ca h s i t m hnh hi quy.
t_critical l gi tr tng ng vi mc  ngha 5% v bc t do.
SE(i) l sai s tiu chun ca c lng i.
Gi tr t_critical c th c tm t bng phn phi Student hoc s dng cc tnh ton thng k trong phn mm h tr.

Lu  rng  tnh ton sai s tiu chun SE(i), chng ta cn s dng thng tin t m hnh hi quy, bao gm sai s tiu chun ca c lng v ma trn hip phng sai ca cc c lng.

V thiu thng tin v cc sai s tiu chun v ma trn hip phng sai trong cu hi ca bn, khng th tnh ton khong tin cy c th. Tuy nhin, phng php v cng thc c cung cp trn s gip bn tnh ton khong tin cy cho 1, 2 v 3 trong trng hp c th ca m hnh hi quy ca bn.

Phi chng c ba bin u khng nh hng n gi PRICE?

 xc nh xem c ba bin SQFT, BEDRMS v BATHS c nh hng n gi PRICE hay khng, chng ta c th s dng cc phng php thng k trong m hnh hi quy nh kim nh gi thuyt.

Trong m hnh hi quy, gi thuyt H0 (gi thuyt khng) l khng c mi quan h tuyn tnh gia bin u vo v bin ph thuc. Ngha l 1 = 2 = 3 = 0.

 kim tra gi thuyt ny, chng ta c th s dng kim nh t-Test cho cc h s 1, 2 v 3. Nu gi tr p (p-value) tng ng vi kim nh t-Test l nh hn mc  ngha  cho (trong trng hp ny, 5%), chng ta c cn c  bc b gi thuyt H0 v kt lun rng c mi quan h tuyn tnh gia bin u vo v bin ph thuc.

Nu gi tr p ln hn hoc bng mc  ngha, chng ta khng c  cn c  bc b gi thuyt H0. iu ny c ngha l khng c bng chng thng k cho thy c ba bin SQFT, BEDRMS v BATHS nh hng n gi PRICE.

V khng c thng tin v gi tr p trong cu hi ca bn, khng th rt ra kt lun chnh xc v tc ng ca c ba bin n gi PRICE. Tuy nhin, bng cch thc hin kim nh t-Test, bn c th nh gi xem c bng chng thng k  ng h hay bc b mi quan h tuyn tnh gia cc bin u vo v gi PRICE.

D bo gi tr trung bnh v gi tr c bit v gi nh trung bnh khi SQRT = 2500, BDRMS = 3 v BATHS = 2?


 d bo gi tr trung bnh v gi tr c bit v gi nh trung bnh khi SQFT = 2500, BEDRMS = 3 v BATHS = 2, chng ta s dng m hnh hi quy  c c lng.

PRICEi = 0 + 1SQFTi + 2BEDRMSi + 3BATHSi + i

u tin, chng ta s dng gi tr ca SQFT, BEDRMS v BATHS  tnh gi tr d bo trung bnh (predicted mean value) ca gi nh (PRICE). iu ny c thc hin bng cch thay th gi tr vo cng thc hi quy:

PRICE_predicted_mean = 0 + 1(SQFT) + 2(BEDRMS) + 3(BATHS)

 bc tip theo, chng ta s dng gi tr ca SQFT, BEDRMS v BATHS  tnh gi tr d bo c bit (predicted individual value) ca gi nh (PRICE) cho trng hp c th ny. Cng thc ny cng s dng cng thc hi quy:

PRICE_predicted_individual = 0 + 1(SQFT) + 2(BEDRMS) + 3(BATHS) + i

Lu  rng  tnh ton gi tr d bo, chng ta cn bit gi tr ca h s hi quy (0, 1, 2, 3). Trong trng hp thiu thng tin v cc h s hi quy, khng th tnh ton gi tr d bo c th. Tuy nhin, phng php v cng thc  c cung cp s gip ta tnh ton gi tr d bo cho SQFT = 2500, BEDRMS = 3 v BATHS = 2 trong trng hp c th ca m hnh hi quy ny.
